```python
batch_size=8
image_generator = ImageDataGenerator(
        samplewise_center=True,
        samplewise_std_normalization= True,
        rotation_range=30,horizontal_flip=True)
generator = image_generator.flow_from_dataframe(
                                                dataframe=finger_study,
                                                directory=image_dir,
                                                x_col='imgpath',
                                                y_col='label',
                                                class_mode="binary",
                                                batch_size=batch_size,
                                                shuffle=True,
                                                seed=1,
                                                target_size=(320,320))
```

```python
raw_train_generator = ImageDataGenerator().flow_from_dataframe(
        dataframe=finger_study, 
        directory=image_dir, 
        x_col="imgpath", 
        y_col='label', 
        class_mode="binary", 
        batch_size=100, 
        shuffle=True, 
        target_size=(320,320))
    
    # get data sample
batch = raw_train_generator.next()
data_sample = batch[0]

    # use sample to fit mean and std for test set generator
image_generator = ImageDataGenerator(
    featurewise_center=True,
    featurewise_std_normalization= True)
image_generator.fit(data_sample)

valid_generator = image_generator.flow_from_dataframe(
        dataframe=valfinger,
        directory=image_dir,
        x_col='imgpath',
        y_col='label',
        class_mode="binary",
        batch_size=batch_size,
        shuffle=False,
        seed=1,
        target_size=(320,320))
```

```python
def getfreqs(generator):
    length=len(generator.labels)
    freqpos=sum(generator.labels)/length
    freqneg=1-freqpos
    return freqpos,freqneg
```

```python
wneg,wpos=getfreqs(generator)
print(wneg,wpos)
```

```python
def weightedbinarycrossentropy(ytrue,ypredicted):
        ytrue=tf.cast(ytrue, tf.float32)
        ypredicted=tf.cast(ypredicted, tf.float32)
        loss=np.array([0.0])
        loss=tf.cast(loss, tf.float32)
        for i in range(len(ytrue)):
            loss+=(-((wpos*ytrue[i]*tf.keras.backend.log(ypredicted[i]))+(wneg*(1-ytrue[i])*tf.keras.backend.log(1-ypredicted[i]))))
        return loss/float(len(ytrue))
```

```python
base_model = DenseNet169(weights='imagenet', include_top=False)

x = base_model.output

# add a global spatial average pooling layer
x = GlobalAveragePooling2D()(x)

# and a logistic layer
predictions = Dense(1, activation="sigmoid")(x)

model = Model(inputs=base_model.input, outputs=predictions)
adam1=tf.keras.optimizers.Adam(learning_rate=0.0001)

model.compile(optimizer=adam1, loss=weightedbinarycrossentropy,metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop=EarlyStopping(monitor='loss',patience=2)
reduce_lr=tf.keras.callbacks.ReduceLROnPlateau(monitor="val_loss",factor=0.1)
```

```python
history = model.fit_generator(generator, 
                              validation_data=valid_generator,  
                              epochs = 10,callbacks=[early_stop,reduce_lr])
```

```python
plt.plot(history.history['loss'])
plt.ylabel("loss")
plt.xlabel("epoch")
plt.title("Training Loss Curve")
plt.show()
```
